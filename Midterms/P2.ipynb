```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

def analyze_transfer_function():
    # 1. Define placeholder numerical values for symbolic variables
    M1 = 1.5   # Mass 1
    M2 = 1.0   # Mass 2
    B1 = 0.8   # Damping coefficient 1
    B2 = 0.4   # Damping coefficient 2
    K1 = 15.0  # Spring constant 1
    K2 = 10.0  # Spring constant 2

    # 2. Define the Transfer Function coefficients
    num = [B1, K1]

    den = [
        M1 * M2,                                      # s^4
        (B1 * M1 + B1 * M2 + B2 * M1),                # s^3
        (B1 * B2 + K1 * M1 + K1 * M2 + K2 * M1),      # s^2
        (B1 * K2 + B2 * K1),                          # s^1
        K1 * K2                                       # s^0
    ]

    # 3. Create the control system object
    sys = ct.TransferFunction(num, den)
    print("Analyzed Transfer Function:")
    print(sys)

    # 4. Generate and plot the Step Response
    plt.figure(figsize=(12, 5))
    
    t, y = ct.step_response(sys)
    plt.subplot(1, 2, 1)
    plt.plot(t, y)
    plt.title('Step Response')
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')
    plt.grid(True)

    # 5. Generate and plot the Bode Plot
    print("Generating Bode Plot...")
    plt.subplot(1, 2, 2)
    ct.bode_plot(sys, dB=True, Hz=True, deg=True, Plot=True)
    
    plt.suptitle('System Analysis: Step Response and Bode Plot')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

if __name__ == "__main__":
    analyze_transfer_function()
```
