```python
import control
import matplotlib.pyplot as plt
import numpy as np

# Define placeholder numerical values for the symbolic variables
# These values are chosen to ensure a stable and illustrative response
M1 = 1.0
M2 = 1.0
B1 = 0.5
B2 = 0.5
K1 = 1.0
K2 = 1.0

# Define the coefficients for the Numerator: (B1*s + K1)
num = [B1, K1]

# Define the coefficients for the Denominator: 
# (M1*M2*s**4 + s**3*(B1*M1 + B1*M2 + B2*M1) + s**2*(B1*B2 + K1*M1 + K1*M2 + K2*M1) + s*(B1*K2 + B2*K1) + K1*K2)
den = [
    M1 * M2,                                   # s^4 coefficient
    (B1 * M1 + B1 * M2 + B2 * M1),             # s^3 coefficient
    (B1 * B2 + K1 * M1 + K1 * M2 + K2 * M1),   # s^2 coefficient
    (B1 * K2 + B2 * K1),                       # s^1 coefficient
    K1 * K2                                    # s^0 coefficient
]

# Create the Transfer Function object
sys = control.TransferFunction(num, den)

print("Transfer Function:")
print(sys)

# 1. Step Response
t, y = control.step_response(sys)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, y)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# 2. Bode Plot
plt.subplot(1, 2, 2)
# control.bode_plot returns magnitude, phase, and frequency
mag, phase, omega = control.bode(sys, plot=True)
plt.gcf().axes[0].set_title('Bode Plot - Magnitude')
plt.gcf().axes[1].set_title('Bode Plot - Phase')

plt.tight_layout()
plt.show()
```
